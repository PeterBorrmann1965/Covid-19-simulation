```python
"""Simulation of infections for different scenarios."""
import os
import covid19sim.coronalib as cl
import pandas as pd 
import numpy as np
import plotly.express as px
import datetime
import plotly.graph_objects as go
from plotly.offline import plot
from plotly.subplots import make_subplots
```

```python
w = 1
age, agegroup, gender, contacts, drate, hnr, persons = cl.makepop("current",83000000/w)
# Einlesen der realisierten Daten
brd = pd.read_excel("./brd_dat.xlsx")
brd["Meldedatum"] = brd["Meldedatum"].dt.date
brd["Datum"] = brd["Meldedatum"]
brd["Fälle"] = brd["Fälle"] / w 
brd["Fälle_kum"] = brd["Fälle_kum"] / w 
```

```python
day0date = datetime.date(2020, 3, 8)
r_change = {}
r_change['2020-01-01'] = 0.4*8 * contacts/np.mean(contacts)
r_change['2020-03-08'] = 0.24*8 * contacts/np.mean(contacts)
r_change['2020-03-16'] = 0.15*8 * contacts/np.mean(contacts)
r_change['2020-03-23'] = 0.6 * contacts/np.mean(contacts)
r_change['2020-05-04'] = 0.7 * contacts/np.mean(contacts)

com_attack_rate = {}
com_attack_rate["2020-01-1"] = 0.0
state, statesum, infections, day0, rnow, args, gr = cl.sim(
        age, drate, nday=180, prob_icu=0.009, day0cumrep=4000/w,
        mean_days_to_icu=16, mean_duration_icu=14,
        mean_time_to_death=21,
        mean_serial=7.5, std_serial=3.0, immunt0=0.0, ifr=0.003,
        long_term_death=False, hnr=None, com_attack_rate=com_attack_rate,
        r_change=r_change, simname="Basis NRW ohne Community Attack",
        datadir=".",
        realized=brd, rep_delay=13, alpha=0.125, day0date=day0date)

cl.plotoverview(gr, args)
```
