# Einleitung

Einfache epidemiologische Modelle wie SIR, SEIR sind sehr hilfreich zum generellen Verständnis einer Pandemie, sind jedoch nicht ausreichend, um die Infektionsdynamik unter sich stetig verändernden Bedingingen zu verstehen und zu progostizieren. Dies gilt insbesondere für das derzeitige Infektionsgeschehen mit dem neuen Sars-Cov-2 und der damit verbundenen Erkrankung Covid-19. Die Modelle sind Makrosimulationen, bei denen die Änderungen der Fallzahlen über die Lösung von Differentialgleichungen berechnet werden. 

Sowohl die relativ lange Inkubationszeit, der hohe Anteil an asymptomatischen und milden Erkrankungsverläufen und die stark alterabhängige Verteilung schwerer und tödlicher Verläufe spielen hier eine wesentliche Rolle. Im Zusammenspiel mit staatlichen Maßnahmen kann es zu paradoxen Situationen kommen, in denen zwar die Reproduktionrate für die Gesamtbevölkerung sinkt, die Anzahl der Intensivfälle und der Toten auch nach Bereinigug durch Effekte von Zeitverzögerungen weiter steigt.

Für die Erklärung des bisherigen Verlaufs der Pandemie und die Prognose des zukünftigen Verlaufs sind zum einen direkt mit dem jeweiligen Virus und dem Krankheitsverlauf - nicht steuerbare - Parameter und steuerbare Variablen, die sich aus dem individuellen Verhalten von Personen und der Güte des Gesundheitsversorgung ergeben.

Krankheitsparameter

* Dauer von der Infektions bis zum Auftreten der ersten Krankheitssymptome (Inkubationszeit)
* Zeitliche Entwicklung der Wahrscheinlichkeit, dass ein Infizierter einen Nicht-Infizierten bei einem "nahen" Kontakt ansteckt (Infektiosität)
* Wahrscheinlichkeit einer schweren Erkrankung mit Intesivbehandlung
* Dauer von der Infektion bis zu Intensivbehandlung
* Dauer von der Infektions bis zur Genesung

Ausbreitungsparameter

* Individuelle ungeschütze Kontaktraten von Nicht-Infizierten mit Infizierten

Prozessuale Parameter 

* Mortalität: Anteil der Infektionen mit letalem Ausgang. Die Mortalität kann von Land zu Land deutlich variieren und ist abhängig von den Kapazitäten des jeweiligen Gesundheitssystems und kann sich durch neue Behandlunsgmethoden und Medikamente im Zeitverlauf deutliche ändern
* Dauer von Hospitalisierungen und Intensivbehandlungen

Alle Parameter können stark vom allgemeinen Gesundheitszustand, dem Alter, dem Geschlecht und spezifischen Vorerkrankungen abhängig sein. 

Für den Vergleich von Prognosen mit gemeldeten Zahlen gibt es weitere Faktoren: 

* Zeitraum zwischen Infektion oder Erkrankung und Meldung
* Vollständigkeit der Erfassung von Infektionen, Erkrankungen, Intensivbehandlungen und Todesfällen
* Bei Todesfälle die verwendete Methode der Zuordnung der Todesursache (an Covid-19 oder mit Covid-9 verstorben)






# Beschreibung des Modells

Im SIR-Modell werden zum Beispiel die zeitlichen Entwicklungen der aktuell infizierten Personen $X(t)$, die Anzahl der empfänglichen/nicht infizierten Personen $S(t)$ und die Anzahl der genesenen Personen $R(t)$ berechnet. Dabei ist $\beta(t)$ die Transmissionrate und $\gamma(t)$ die Genesungsrate. 
\begin{align*}
\frac{dS(t)}{dt} &= \frac{-\beta (t) S(t)X(t)}{n},\label{diff_time_sir1}\\
\frac{dX(t)}{dt} &= \frac{\beta (t) S(t)X(t)}{n} -\gamma (t) X(t),\label{diff_time_sir2}\\
\frac{dR(t)}{dt} &= \gamma (t) X(t).\label{diff_time_sir3}
\end{align*}


Unser Modell ist eine Mikrosimulation, d.h. jeder Person wird individuell simuliert. Auf diese Weise können unterschiedliche Kontaktraten und Sterbewahrscheinlichkeiten bei Infektion berücksichtigt werden. 

0. Erstellung einer Population mit der gewünschten Größe und den gewünschten demographischen Merkmalen
1. Inititalisierung der Pandemie mit einer geringen Zahl von anfänglich Infizierten
2. Simulation der Entwicklung der Infizierten, Immunisierten/Genesenen, der ICU-Fälle und der Toten über einen Zeitraum von k Tagen
    * Berechnung der erwarteten Neuinfektionen für den Tag t 
    * Zufällige Ziehung neuer Infizierter gemäß ihrer relativen Infektionswahrscheinlichkeit.
    * Zufällige Zuweisung des Status "Intensiv" in Abhängigkeit von der individuellen Wahrscheinlichkeit einer Intensivbehandlung und der Tage seit Infektion. Gleichzeitig wird eine individuelle Verweildauer zufällig gemäß einer Verteilungsfunktion ermittelt.
    * Zufällige Zuweisung des Status "verstorben" in Abhängigkeit von der individuellen Wahrscheinlichkeit an einer Covid-19 Infektion zu sterben und der Tage seit Infektion 
    * Nicht verstorbene und nicht intensivbehandelte Infizierte werden 28 Tage nach Infektion auf den Status genesen gesetzt
    
Die einzelnen Schritte werden im Folgenden näher erläutert. Grundsätzlich können weitere Zwischenzustände "symptomatisch" oder "hospitalisiert" verwendet werden. Zur Ermittlung der zusätzlichen Toten können individuelle Sterbewahrscheinlichkeiten herangezogen werden. 


## Standardpopulationen
Da die Infected Fatality Rate (IFR) stark vom Alter eines Infizierten abhängt, ist es wesentlich eine repräsentative Population in den Simulationen zu verwenden. 

Eine Population besteht aus n Personen, für die jeweils folgende Eigenschaften vorhanden sind: 


* Alter
* Geschlecht
* Familienstand
* Mittlere Kontaktrate
* Sterbewahrscheinlichkeit (1 Jahr)
* Lebenserwartung in Jahren
* Haushaltsnummer (nur Stichprobe 2010)
* Anzahl weiterer Personen im Haushalt oder der Gemeinschaftsunterkunft (nur Stichprobe 2010)
    
Für die Struktur der Basispopulation werden zwei unterschiedliche Verfahren verwendet: 


1. Population2019: Die Struktur entspricht der Fortschreibung des Bevölkerungsstandes nach Alter, Geschlecht und Familienstand gemäß dem Statistischen Bundesamt für Anfang 2019, d.h. die Anteile jeder Gruppe entsprechen genau denen der aktuellen Bevölkerungsstruktur.
2. Stichprobe2010: Die Basispopulation ist gleich der Population aus dem Campusfile des Statistischen Bundesamt für den Mikrozensus 2010. Diese Daten enthalten pro Person auch zusätzlich eine Haushaltsnummer und die Anzahl der weiteren Personen im Haushalt. Das Campus-File enthält xxx Personen verteilt auf yyy Haushalte. Für die Simulation wird dies Population so oft repliziert bis die gewünschte Populationsgröße erreicht wird. Jede Replik erhält dabei neue Haushaltsnummern.


Der Vorteil der Population2019 ist, dass diese eine hohe Repräsentativität  für den aktuellen Bevölkerungsstand hat und mit Alter und Geschlecht zwei wesentliche bekannte Einflussfaktoren der Sterblichkeit durch Covid-19 enthält.

## Modellparameter


1. Die Sterbewahrscheinlichkeit bei Infektion ist proportional zur langfristigen Sterbewahrscheinlichkeit gemäß den Sterbetafeln des statistischen Bundesamtes unter Berücksichtigung von Alter und Geschlecht. 
2. Die individuellen Kontaktraten sind Altersgruppen-spezifisch. (https://journals.plos.org/plosmedicine/article/file?id=10.1371/journal.pmed.0050074&type=printable)

Wesentlich für die Infektionsdynamik ist das Infektionsprofil, das angibt, wieviele Menschen ein Infizierter im Mittel in Abhängigkeit der Tage seit seiner eigenen Infektions infiziert. Dieses Profil ist zum einem vom Krankheitsverlauf und dem damit verbundenen "Ausstoß" aktiver Viren, zum anderen vom individuellen Verhalten des Infizierten und den getroffenen Schutzmaßnahmen abhängig. 
![Infektionsprofil](scripts/pdf.png)
Die Abbildung zeigt drei Beispiele. "Ohne Maßnahmen" ist das typische mittlere Profil, wenn keine Maßnahmen getroffen werden. Eine Isolation von 50% der Infizierten nach 4 Tagen, halbiert ab Tag 5 die mittleren Infektionsrtaten. Bei der Kurve "Lockdown" wurde angenommen, dass die Anzahl der Kontakte generell um einen Faktor 3 gesenkt wurde. Die Fläche unter den Kurven ist jeweils die <b>effektive Reproduktionsrate (Re)</b>. Im Beispiel ergeben sich Werte von 2,5 (ohne Maßnahmen), 0.9 (Lockdown) und 1,6 (Isolation). 

Für zukünftige Prognosen ist wesentlich, die Effekte von Maßnahmen auf die individuellen Infektionsprofile abzuschätzen. Beispiele: 


* Spezifische Kontaktbeschränkungen unterschiedlicher Berufs- und Bevölkerungsgruppen müssen berücksichtigt werden. 
* Hinsichtlich des Trages von Masken muss nicht nur eingeschätzt werden, wie stark sich das Übertragungsrisiko pro  Kontakt reduziert, sondern auch abgeschätzt werden, wie konsequent Masken in unterschiedlichen Bevölkerungs- und Berufsgruppen getragen werden - und in welchem Maße Masken verfügbar sind. 
* Bei einer freiwillig genutzten Corona-App ist neben der Durchdringung vor allem die Disziplin bei der Selbstisolation entscheidend. In der Theorie kann eine solche App große Effekte haben, da nach einer festgestellten Neuinfektion nicht nur die direkten sondern auch die indirekten Kontakte informiert werden könnten. 


Die <b>effektiven Reproduktionsraten</b> vor und nach dem Lockdown sind die Simulationsparameter mit dem stärksten Effekt auf die Infektionsdynamik. In der Simulation wird die mittlere Repoduktionsrate über die Gesamtpopulation eingestellt. Die individuelle Infektionswahrscheinlichkeit ist proportional zu individuellen Kontaktrate. Bei gleichem mittlerem Re haben die Kontaktraten keine Auswirkung auf die Entwicklung der Gesamtzahl der Infizierten aber auf die Anteile der Infizierten in den unterschiedlichen Altersgruppen. Derzeit sind ältere Menschen in Deutschland noch unterdruchschnittlich oft infiziert. Für mittelfristige Simulationen ist zu erwarten, dass die mittleren Kontaktraten sich abhängig von den jeweiligen beschlossenen Maßnahmen in den Altersgruppen sehr unterschiedlich entwickeln können. 

Die <b>Infected Fatality Rate (IFR)</b> gibt das mittlere Verhältnis zwischen Toten und Infizierten nach Fallabschluss an. Das die Sterberate bei Infektions stark vom Gesundeheitszustand und dem Alter der Infizierten abhängt, hängt die IFR stark von der Zusammensetzung der Infizierten ab. In unserer Simulation wird die IFR so eingestellt, dass sie der mittlereren Sterbewahrscheinlichkeit bei Infektion über die gesamte Bevölkerung entspricht. Die indivuelle IFR ist dabei proportional zur Sterbewahrscheinlichkeit auf Basis der Sterbetafeln des Statistischen Bundesamtes. In der Simulation kann die gemessene IFR niedriger oder höher ausfallen, wenn ältere Menschen unter- oder überdurchschnittlich infiziert sind. Wenn alle Menschen infiziert sind, stellt sich genau die als Simulationsparameter verwendete IFR ein. 

Für die Abschätzung der IFR aus den gemeldeten Fällen und den gemeldeten Toten müssen mehrere Faktoren berücksichtigt werden. 

* Zwischen Infektionszeitpunkt und und Meldezeitpunkt (Tag des Tests) liegen in der Regel mehrere Tage.
* Nur ein Teil der Infizierten wird durch einen Test erkannt. 
* Zwischen Infektionszeitpunkt und Todeszeitpunkt gibt es eine erhebliche Verzögerung von 15-25 Tagen
* Es können in Deutschland mehrere Tage (bis Wochen) vergehen bis ein Infektionsfall oder Todesfall in der Statistik erscheint, d.h. die Todesfälle für einen bestimmten Tag steigen rückwirkend im Zeitverlauf an. 

Die naive Abschätzung der IFR als Quotient der kumulierten Toten und der kumulierten gemeldeten Infizierten ist grob falsch. Eine deutlich besser Abschätzung ergibt sich unter Berücksichtigung des Zeitverzugs zwischen Meldung der Infektion und Meldung des Todes. Zwischen den neu Infizierten I(t) und den gemeldeten Fällen besteht die Beziehung: 
\begin{equation*}
R(t) = \gamma \sum_{s=0}^t  \mathrm{I}(t-s) \; \mathrm{d}_\mathrm{IR}(s)
\end{equation*}
Dabei gibt $\gamma$<1 den Anteil der Infizierten an, die identifiziert werden. Die Funktion 
$\mathrm{d}_\mathrm{IR}(s)$ beschreibt die Wahrscheinlichkeit, mit welchem Zeitverzug ein Infizierter getestet wird.
Für die Anzahl der Toten T(t) am Tag t kann man annehmen: 
\begin{equation*}
T(t) = \mathrm{IFR} * \sum_{s=0}^t \mathrm{I}(t-s) \; \mathrm{d}_\mathrm{IT}(s)
\end{equation*}
wobei $\mathrm{d}_\mathrm{IT}(s)$ bei Sterbefällen die Wahrscheinlichkeit angibt, dass der Tod s Tage nach der Infektion eintritt. 
\begin{equation*}
    T(t) = \mathrm{CFR} * \sum_{s=0}^t \mathrm{R}(t-s) \; \mathrm{d}_\mathrm{RT}(s)
\end{equation*}
Hierbei ist die CFR die Case Fatality Rate, d.h. die Sterblichkeit bezogen auf die gemeldeten Fälle. Die Funktion $\mathrm{d}_\mathrm{RT}(s)$ gibt bei Sterbefällen die Wahrscheinlichkeit an, dass der Tod s Tage nach der Infektion eintritt. 

Zwischen CFR und IFR besteht die Beziehung.
\begin{equation*}
\mathrm{CFR} = \gamma \mathrm{IFR}
\end{equation*}
Zu beachten ist hierbei, dass $\gamma$ im Zeitverlauf varieren kann. Z.B. sinkt $\gamma$ unweigerlich, wenn die Testkapazitäten erschöpft sind.

Die Funktionen $\mathrm{d}_\mathrm{IR}(s)$, $\mathrm{d}_\mathrm{IT}(s)$ und  $\mathrm{d}_\mathrm{RT}(s)$ können jeweils durch einen mittleren Zeitverzug und eine Standardabweichung charakterisiert werden. Die Werte hierfür können bezogen auf den den Infektionszeitpunkt (IR und IT) jeweils nur in Einzelfällen, wenn bekannt ist, wann der Kontakt zum Überträger der Infektion stattfand, nachvollzogen werden.

Die Funktion $\mathrm{d}_\mathrm{RT}(s)$ ist länderspezifisch und kann sich im Zeitverlauf ändern, könnte jedoch grundsätzlich gut aus den Meldedaten ermittelt werden, da pro Todesfall die Zeitdifferenz zwischen Positivtestung und Eintritt des Todes gemeldet wird. In Deutschland hat das RKI hierzu noch keine Zahlen veröffentlicht. 

Qualitativ kann man sagen, dass in Ländern wie Südkorea, die umfangreiche und freiwillige Tests auch bei vollkommen symptomfreien Personen durchführen, die mittlere Zeit zwischen Infektion und Tod deutlich größer ist als in Ländern wie in Italien, in denen nur Personen mit mittleren bis schweren Symptomen gestestet werden oder der Test sogar post mortem (Zeitverzug 0) druchgeführt wird.

Der Wert von $\gamma$ ist ebenfalls länderspezifisch. Schätzungen reichen hier 1/2 bis zu 1/50. # Reproduktionsrate R0 vor dem Lockdown

## Datenlage in Deutschland
Eine exakte Messung der Reproduktionsrate R0 ist unseres Erachtens nicht möglich. 



Davon abgesehen spielen weitere Faktoren eine Rolle: 

* Für Covid-19-Meldefälle, Intensiv-Fälle und Todesfälle gibt es Zeitverzüge zum 
    Infektionszeitpunkt von einer bis mehreren Wochen. Derzeitige Anstiege spiegeln also das 
    Infektionsgeschehen von vor Tagen bzw. vor Wochen wieder. 
* Die Meldefälle hängen von den Testvorgaben (mit Symptomen und Kontakt) und den jeweiligen
    Testkapazitäten ab. Bei exponentiell steigenden Infektionsfällen ist anzunehmen, 
    dass die Dunkelziffer im Zeitverlauf steigt. 
* Aufgrund der auf die Gesamtbevölkerung bezogen geringen Anzahl an Toten pro Tag gibt es ein großes statistisches Rauschen und zufällige Ereignisse (Z.B. kann ein einziges betroffenes Altenheim derzeit eine große Auswirkung auf die Todesfälle in einem Landkreis oder sogar Bundesland haben.).



Für die <b>Basisreproduktionsrate R0 in Deutschland vor dem Lockdown sehen wir einen Bereich von 2.0 bis 2.5 </b> als realistisch an. Hierbei ist bereits eingeflossen, dass bereits vor dem Lockdown Schulen geschlossen wurden, Großveranstaltungen weitestgehend untersagt wurden und die allgemeine Sensibilisierung der Bevölkerung für Hygiene und Social Distancing bereits eingeleitet war.

## Vergleich mit asiatischen Ländern
Für einen Vergleich der Kontaktraten unter Normalbedingungen liegen uns keine Studiendaten, die einen direkten Vergleich ermöglichen, vor. Es ist jedoch davon auszugehen, dass auch unter Normalbedigungen in asiatischen Ländern wie Japan oder
Südkorea eher kontaktarme Begrüßungsrituale üblich sind und das Tragen vom Masken in dichten Menschenmengen sehr häufig praktiziert wird - nicht nur um sich selbst, sondern auch um andere Menschen zu schützen. <br>
Unser Einschätzung ist: <b>Die Basisreproduktionsrate R0 in Deutschland ist höher als in Ländern wie Japan oder
Südkorea.</b>

# Einschätzung der Reproduktionsrate Re nach dem Lockdown

## Datenlage in Deutschland

Die aktuell fallenden Zahlen von Neu-Infektionen deuten darauf hin, dass die effektive Reproduktionsrate 
unter Lockdown-Bedingungen in Deutschland unter 1 gefallen sein könnte. Aufgrund möglicher Verzerrungen (limitierte Anzahl 
an Tests, verstärkte Testungen von systemrelevanten Bevölkerungsteilen, ...) ist aber auch eine effektive Reproduktionsrate 
größer als 1 nicht komplett auszuschließen. 

Wegen der kürzeren Inkubationszeit der Influenza müssen bei einem wirksamen Lockdown die gemeldeten Influenza-Infektionen früher zurückgehen als die Covid-19 Fälle. Laut Bericht des RKI zur KW 13 (21.-27.3.2020) ist dies der Fall (https://influenza.rki.de/Wochenberichte/2019_2020/2020-13.pdf). Die Zahl der gemeldeten Fälle ist in den letzten beiden Wochen stark gesunken: 19.130 (KW11), 10.712 (KW12), 3.528 (KW13). Wegen des saisonal bedingten Abklingens der Grippewelle kann daraus jedoch nur geschlossen werden, dass die Grippemeldungen einer deutlich Veringerung des allgemeinen Infektionsgeschehens nicht entgegenstehen. 

In Großbritanien wurde der Effekt des Lockdowns auf eine Reduktion der Kontakte um ca. 75% geschätzt 
(https://cmmid.github.io/topics/covid19/current-patterns-transmission/comix-impact-of-physical-distance-measures-on-transmission-in-the-UK.html)
Es ist zu beachten, dass sich relative Reduktionen nicht 1:1 übertragen lassen und die Kontaktraten unter

## Lockdown-Effekt in Österreich
Österreich hat bereits am 13.3.2020 einen Lockdown beschlossen. Zu diesem Zeitpunkt gab es in Österreich einen einzigen bestätigten Covid-19 Todesfall. Die Maßnahmen werden strikter als in Deutschland eingeschätzt. Zudem scheint die Versorgung mit Atemschutzmasken besser als in Deutschland zu sein. In Österreich zeigen sich bereits deutliche Effekte des Lockdowns:

* Die Zahl der gemeldeten täglichen Neuinfektionen sinkt seit dem 28.3.2020 (15 Tage nach Lockdown) von einem Maximum von 1043 am 27.3. auf 282 am 4.4.2020, wobei hier noch Nachmeldungen zu erwarten sind. 
* Die Zahl der gemeldeten Toten ist seit dem 31.3 (22 Tote) nicht mehr ansteigend.

## Lockdown-Effekt in Italien
In Italien wurde der Lockdown am 09.03.2020 eingeführt. Tests auf Covid-19 werden in Italien erst bei mittleren bis schweren Symptomen oder sogar post mortem durchgeführt. Wegen begrenzter Testkapzitäten bei gleichzeitig hoher Dunkelziffer ist eine gleichbleibende Anzahl an gemeldeten Neuinfektionen noch kein eindeutiges Indiz dafür, dass der exponentielle Anstieg gestoppt wurde. <br> 
Auch die Anzahl der Intensivpatienten kann nicht verlässlich zur Einschätzung herangezogen werden, da in einigen Regionen die Kapazitätsgrenzen überschritten wurden und selbst bei sinkenden Zahlen im Bedarf weiterhin ein Ausbau von Kapazitäten notwendig ist.<br>
Als erstes positives Zeichen kann gesehen werden, dass die Zahl der Covid-19-Toten seit einigen Tagen sowohl in der am stärksten betroffenen Region Lombardei als auch in ganz Italien sinkt.


# Simulationsergebnisse
Die folgende Tabelle zeigt abhängig vom Szenario die benötigten ICU-Kapazitäten als Vielfaches der zum Lockdown belegten Kapazitäten. Lesebeispiel: Bei einer Belegung von 500 ICU-Betten zum Lockdowh-Zeitpunkt würden im Szenario R0=3,5 und Rlock=0,9 maximal 5950 Betten (11,9 x 500) 25 Tage nach dem Lockdown notwendig. 

| R0  | Rlock | Peak Tag | Peak/ICU Lock |
|-----|-------|----------|---------------|
| 2,5 | 0,7   | 20       | 4,7           |
| 2,5 | 0,9   | 25       | 7,2           |
| 2,5 | 1,2   | 112      | 29,5          |
| 3,5 | 0,7   | 20       | 9,0           |
| 3,5 | 0,9   | 25       | 11,9          |
| 3,5 | 1,2   | 99       | 33,8          |

R0 hat einen starken Einfluß auf die benötigte Bettenzahl, wenn Rlock kleiner als 1 ist. Hier kommt der Effekt zum Tragen, dass zum Lockdown-Zeitpunkt bereits deutlich mehr Infizierte vorhanden sind, die in den Folgewochen versorgt werden müssen. 

Die Lockdown-Werte von 0,7 und 0,9 führen zwar immer noch zu einem kurzfristigen Anstieg des ICU Bedarfs um einen Faktor 4,7-11,9, aber vermutlich nicht zu einer Überlastung der Gesamt-Kapazitäten. 

Die beiden Szenarien mit Rlock=1.2 sagen eine sehr hohe Maximalbelastung nach erst 3-4 Monaten vorher. Hier bestünde die Herausforderung, die effektive Reproduktionszahl weiter zu senken und gleichzeitig die wirtschaftliche Produktivität wieder zu steigern. 

Zusammenfassend kann davon ausgegangen werden, dass Deutschland kurzfristig unter Lockdown-Bedingungen an einer Überlastung des Gesundheitssystems vorbeikommen könnte. Es kommen hier im Vergleich zu anderen europäischen Ländern mehrere positive Effekte zusammen: 
    
    
* Vergleichweise **geringe Kontaktraten in Deutschland** führen zu einer geringeren Basisreproduktionszahl. 
* Durch **umfangreichere Tests** konnte gerade in der Anfangsphase ein besseres Containment durchgeführt werden. Auch derzeit sind die Positivfälle in vielen Kommunen noch so gering, dass eine frühzeitige Identifikation und Isolation von Kontaktpersonen möglich ist. 
* Deutschland hat im Vergleich zu vielen anderen Ländern **früher umfassende Kontaktsperren** eingeführt. Bei einem Lockdown unter Re=1 kann bereits eine Woche längeres Warten einen Faktor 3 in der Maximalbelastung der Intensivstationen bedeuten (In der Lombardei waren zum Zeitpunkt des Lockdowns bereits 440 Covid-19 Patienten in Intensivbehandlung. Eine Woche zuvor waren es nur 127).
* Die **Intensivkapazitäten sind in Deutschland vergleichsweise hoch** - jedoch verschafft auch eine 3-fache Kapazität bei einem ungebremsten Wachstum an Infizierten von z.B. 20% pro Tag nur einen zusätzlichen zeitlichen Puffer von ca. 6 Tagen.


Anmerkung: In Deutschland befindet sich ein Intensivregister zu Covid-19 Patienten im Aufbau (https://www.divi.de/register/kartenansicht). Da sich die Fallzahlen jedoch sowohl durch zusätzliche Patienten als auch durch neu hinzukommende Intensivstationen, die Covid-19-Fälle melden, erhöhen, ist die Zeitentwicklung nicht zuverlässlich einschätzbar.
