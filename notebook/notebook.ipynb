# General Description of our Project

The problem we are concerend with is to simulate possible scenarios for the further spreading of the new Covid-19 virus
, which originated in Chinas Wuhan province December 2019 and has now succesfully swamped over to great parts of the world. Since march, espescially the western world has become the main affected area  whereas China has succesfully stopped the further spreading of the virus. 

In this study we mainly focus on the simulation of the impact on Germany, but if comparable data for other countries are available, the same methology can be applied driectly.

The main advantage of our approach is that we simulate every individual seperatly whereas standard approaches are only concerend with much groaser levels of aggregation (cite!). 
This takes allows us to analyze the impact Corona on a much finer level then it is, to our best knowledge, currently possible.

As an example, we are able to simulate the effect of how people are distributed on households in every agegroup. This is relevant to see effects like how rest homes affect the the infection and death profile in the risk group of elderly people.



# Description of input Data


Our input data consits of the average contact rate per age group (Mossong et al. 2008) 
and the average mortality rate per age group (destatis). We furthermore leverage a third datafile which contains information of how many people life in a household per age group, which is at the moment not public available (Campusfile). 
To get a feeling for the input data we plot a few distributions below.


```python
fig = go.Figure()
# Change the bar mode
fig = make_subplots(rows=1, cols=2, subplot_titles=("Contact rate over age", "Death rate by age")
                    ,horizontal_spacing=0.15)
fig.add_trace(go.Scatter(x=popi.age, y=popi.contacts_mean,
                    mode='markers', name='contact rate over age'),row=1,col=1)   
fig.add_trace(go.Scatter(x=popi.age, y=popi.loc[(popi.gender=='m'),:].deathrate,
                    mode='markers', name='death rate over age: males'),row=1,col=2)    
fig.add_trace(go.Scatter(x=popi.age, y=popi.loc[(popi.gender=='f'),:].deathrate,
                    mode='markers', name='death rate over age: females'),row=1,col=2) 

fig['layout']['xaxis1'].update(title='age')

fig['layout']['xaxis2'].update(title='age')


fig['layout']['yaxis1'].update(title='contact rate')

fig['layout']['yaxis2'].update(title='death rate')

fig.show()
```

Onthe left we see that the contact rate shows the expected behaviour. It first increaser with age when children get to children garten age and even further when they start attending school. It then drops to a more or less constant rate when peope are in their working life and further decreases when they drop out into retirement. On this scale we can't see the specific effect of the housing environment (familiysize, retirement homes...). We come back to this point later.

On the right we observe the expected behaviour. The deathrate increases with age and for older people men have a significatally higher deathrate then women

```python
popi_with_households=pd.read_csv(pathp+"/population_campus.csv",sep=',')
```

```python
mean_pers_household=popi_with_households.groupby(['age']).Personenzahl.mean().to_frame().reset_index()
mean_contact_number=popi.groupby(['age']).contacts_mean.mean().to_frame().reset_index()
```

```python
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=mean_pers_household.age, y=mean_pers_household.Personenzahl,
                    mode='markers', name='mean number of persons in household'))    
fig2.add_trace(go.Scatter(x=mean_contact_number.age, y=mean_contact_number.contacts_mean,
                    mode='markers', name='mean number of contacts')) 
fig2.update_layout(title="Compare number of contacts with number of persons in household",
    xaxis_title="Age",
    yaxis_title="Mean")
fig2.show()
```

We observe the for elderly people over which are especially vunerable to Covid-19, a great ratio of their contacts is 
created by their living environment. This makes isolation attempts additionaly hard for those age segment.
